```python
import cv2
from cv2_plt_imshow import cv2_plt_imshow
import matplotlib.pyplot as plt
```

# Loading and Plotting Original Image

```python
image = cv2.imread("wallpaper.jpg")
img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
```

# Converting Image to GrayScale

```python
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(grey_image,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
```

#  Inverting the Image

```python
inverted_image = cv2.bitwise_not(grey_image)
plt.figure(figsize=(8,8))
plt.imshow(inverted_image,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
```

```python
image_smoothing = cv2.GaussianBlur(inverted_image, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(image_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
```

# Converting your images to pencil sketches

```python
final = cv2.divide(grey_image, 255 - image_smoothing, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final/Pencil Sketch Image")
```

# Final Output

```python
plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(image)
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(grey_image,cmap="gray")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(inverted_image,cmap="gray")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(image_smoothing,cmap="gray")
plt.title("Smoothen Image")
plt.subplot(1,5,5)
plt.imshow(final,cmap="gray")
plt.title("Final Sketch Image")
plt.show()
```
